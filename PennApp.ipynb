```python
'''Credit to Muhammad Moin for setting up and installing hugging face and llama 2 LLM model

https://github.com/MuhammadMoinFaisal/LargeLanguageModelsProjects/tree/main/Run%20Llama2%20Google%20Colab

'''
model_name_or_path = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q5_1.bin"
```

```python
from huggingface_hub import hf_hub_download
```

```python
from llama_cpp import Llama
```

```python
model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
```

```python
# Have to use GPU
lcpp_llm = None
lcpp_llm = Llama(
    model_path=model_path,
    n_threads=2,
    n_batch=512,
    n_gpu_layers=32
    )
```

```python
prompt = "What is PennApps"
prompt_template=f'''
Your Question: {prompt}

'''
```

```python
from gtts import gTTS
from IPython.display import Audio

loading = 'Please give me a moment as I find the answer to your question'
tts = gTTS(loading)
tts.save('loading.wav')
sound_file = 'loading.wav'
Audio(sound_file, autoplay=True)

```

```python
#generating response might take any where from sec to mins
response=lcpp_llm(prompt=prompt_template, max_tokens=512, temperature=0.5, top_p=0.95,
                  repeat_penalty=1.2, top_k=150,
                  echo=True)
#check for response
print(response["choices"][0]["text"])
resp = response["choices"][0]["text"]
tts = gTTS(resp)
tts.save('resp.wav')
sound_file = 'resp.wav'
Audio(sound_file, autoplay=True)
```
